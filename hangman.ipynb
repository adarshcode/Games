```python
import random as rn
from collections import Counter as cn

swrds='''apple,banana,strawberry,mango,guava,papaya,pineapple,grapes,orange'''
swrds=swrds.split(',')
wrd=rn.choice(swrds)

print("GUESS THE WORD!!")
print("HINT:- IT IS A NAME OF A FRUIT!!")

for i in wrd:
    print("_",end=' ')
print()
print("THE LENGTH OF THE WORD IS",end=' ')
print(len(wrd))

guessedletter=''
chances=len(wrd)+3
print()
print("Chances:",end=' ')
print(chances)
crct=True
flag=True

while(chances!=0) and flag==True:
    print()
    chances-=1
    gwrd=str(input("Guess a letter: "))
    gwrd=gwrd.lower()
    
    if not gwrd.isalpha():
        print("Enter only letters")
        continue
    elif len(gwrd)>1:
        print("Enter only a single letter")
        continue
    elif gwrd in guessedletter:
        print("You have already guessed the letter!!")
        print("Enter another word")
        continue
        
    if gwrd in wrd:
        cnt=wrd.count(gwrd)
        for i in range(cnt):
            guessedletter+=gwrd
    
    for letter in wrd:
        if letter in guessedletter and (cn(guessedletter)!=cn(wrd)):
            print(letter,end=' ')
            crct=False
        elif (cn(guessedletter)==cn(wrd)):
            print()
            print("The word is:",end=' ')
            print(wrd)
            flag=False
            print("CONGRATULATIONS, YOU WON!!")
            break
        else:
            print('_',end=' ')
    
    if flag==False:
        print()
        print("YOU ARE AWESOME!")
        print("BUT THE CREATOR OF THIS PROGRAMME IS KING OF AWESOMNESS!!")
        print("LOL!!!")
    else:
        print()
        print("Chances left:",end=' ')
        print(chances)
            

                
if chances<=0 and (cn(guessedletter)!=cn(wrd)):
    print()
    print("YOU LOST!! TRY AGAIN.....")
    print("The word was: ",end=' ')
    print(wrd)
```
